```python
import transformers
import torch

model_id = "/workspace/model/Meta-Llama-3.1-8B-Instruct"

pipeline = transformers.pipeline(
    "text-generation",
    model=model_id,
    model_kwargs={"torch_dtype": torch.bfloat16},  
    device_map="auto",  
)

messages = [
    {"role": "system", "content": "kamu adalah asisten yang membantu menjawab pertanyaan dengan bahasa yang mudah dipahami."},
    {"role": "user", "content": "Apa yang dimaksud dengan Kewenangan Desa?"},
]

outputs = pipeline(
    messages,
    max_new_tokens=256,
)
print(outputs[0]["generated_text"][-1])
```
