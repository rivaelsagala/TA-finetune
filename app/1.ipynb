```python
# Import library yang diperlukan
import pandas as pd

# Membaca data CSV
data = pd.read_csv('data/data.csv')
print("Data berhasil dimuat!")
print(f"Jumlah data: {len(data)} baris")
print(f"Kolom: {list(data.columns)}")
```

```python
# Melihat data
print("5 data pertama:")
print(data.head())
print("\nInfo data:")
print(data.info())
```

```python
# Analisis sederhana
print("RINGKASAN DATA:")
print(f"Rata-rata umur: {data['Age'].mean():.1f} tahun")
print(f"Rata-rata gaji: Rp {data['Salary'].mean():,.0f}")
print(f"Gaji tertinggi: Rp {data['Salary'].max():,.0f}")
print(f"Gaji terendah: Rp {data['Salary'].min():,.0f}")

print("\nJumlah karyawan per kota:")
print(data['City'].value_counts())
```

```python
# Filter data berdasarkan kriteria
print("KARYAWAN DENGAN GAJI > 6 JUTA:")
gaji_tinggi = data[data['Salary'] > 6000000]
print(gaji_tinggi[['Name', 'City', 'Salary']])

print("\nKARYAWAN DARI JAKARTA:")
jakarta = data[data['City'] == 'Jakarta']
print(jakarta[['Name', 'Age', 'Salary']])
```

```python
# Menambah data baru
data_baru = {
    'ID': 11,
    'Name': 'Fitri Handayani', 
    'Age': 28,
    'City': 'Bogor',
    'Salary': 5500000
}

# Menambahkan ke dataframe
data_lengkap = pd.concat([data, pd.DataFrame([data_baru])], ignore_index=True)
print("Data baru berhasil ditambahkan!")
print(f"Total data sekarang: {len(data_lengkap)} baris")
print("\nData terakhir:")
print(data_lengkap.tail())
```

```python
# Menyimpan data yang sudah diubah
data_lengkap.to_csv('data/data_updated.csv', index=False)
print("Data berhasil disimpan ke 'data/data_updated.csv'")

# Menampilkan ringkasan akhir
print("\nRINGKASAN AKHIR:")
print(f"Total karyawan: {len(data_lengkap)}")
print(f"Rata-rata gaji: Rp {data_lengkap['Salary'].mean():,.0f}")
print(f"Kota terbanyak: {data_lengkap['City'].mode().iloc[0]}")
```
